# 本节重点
- 原函数为$y=f(x)$则其逆函数为$x=f^{-1}(y)$
- 原函数为$y=e^x$则其逆函数为$x=ln(y)$

- 对数函数的性质1:$ln(y*Y) = ln(y)+ln(Y)$
>- $令y = e^x;Y=e^X;y*Y = e^x*e^X;两边取对数:ln(y*Y) = lne^{(x+X)}=x+X=lny+lnY$
- 对数函数的性质2:$ln(y^n) = nln(y)$
>- $令y=e^x;y^2=e^x*e^x;两边同时取对数:ln(y^2) = lne^{(x+x)}=x+x=lny+lny=2lny$

- 例子1
>- $y=x^2$ 逆函数为$x=\sqrt{y}$,注意，这个地方y是不能为负数的，只研究y为正的情况，y为负的情况不是研究范畴

- 例子2
>- $A = \pi r^2 = f(r)$逆函数为$r=\sqrt{\frac{A}{\pi}}=f^{-1}(A)$

- $lny的导数\frac{d(lnx)}{dx}=\frac{1}{x}$

# 函数与逆函数的图像

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
r1 = np.arange(0,1,0.1)
y1 = math.pi*r1**2
```

```python
y2 = np.arange(0,math.pi, 0.1)
r2 = np.sqrt(y2/math.pi)
```

```python
fx, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axs[0].plot(r1, y1)
axs[1].plot(y2, r2)
```

- 可以看出原函数与逆函数的图像是沿y=x这条线对称的

# 指数函数与对数函数的图像
- $ln1=0$
- $lne^1 = 1$
- $ln\frac{1}{e}=ln e^{-1}=-1$

```python
x = np.arange(-2,1.8,0.1)
y = np.exp(x)

x_1 = np.arange(0,5,0.1)
y_1 = np.log(x_1)
```

```python
x_axis = np.arange(-5,5,0.1)
y_axis = np.zeros_like(x_axis)

y_axis_1 = np.arange(-5,5,0.1)
x_axis_1 = np.zeros_like(y_axis_1)
plt.plot(x_axis,y_axis,x_axis_1,y_axis_1,x_1,y_1,x,y)

plt.text(2,4,"y=e^x")
plt.text(3,2,"y=lnx")
```
