- 应用：上班的最短时间（求最小值）

## 二阶导数的作用
- 在求函数极大值极小值时，一阶导数很有用，可以通过令导数=0找到极值点
- 通过一阶导数找到极值点后，可以通过二阶导数得到这个极值点是极大值还是极小值，二阶导数表明函数是朝上还是朝下弯曲,他的符号表示了曲线弯曲的方向，如果为正，表示原始函数的斜率正在增加，为向上弯曲，原始函数为凸函数，反之为减小，原始函数为凹函数
- 二阶导数为0的点为函数的拐点，在这一点，函数的凹凸性质会发生改变。二阶导数小于0表示斜率持续降低，二阶导数大于0表示斜率持续上升。

## 各个函数的表示
- 原始函数：$y=x^2$（可以看成是距离）
- 原始函数的一阶导数：$\frac{dy}{dx}=2x$（可以看成是速度）
- 原始函数的二阶导数：$\frac{d^2y}{dx^2}=2$（可以看成是加速度）

```python
fig,axs = plt.subplots(ncols=3, nrows=1, figsize=(12,4))
x = np.arange(1,100,1)
y = x**2
axs[0].plot(x,y)
axs[0].text(60,6000,"y=x²",size=15)

x_1 = np.arange(1,100,1)
y_1 = 2*x
axs[1].plot(x_1, y_1)
axs[1].text(60,120,"dy/dx=2x",size=15)

x_2 = np.arange(1,100,1)
y_2 = np.zeros_like(x_2)+2
axs[2].plot(x_2, y_2)
axs[2].text(60,2,"d²y/dx²=2",size=15)
```

- 原函数$y=sin(x)$
- 原函数的一阶导数$y=cos(x)$
- 原函数的二阶导数$y=-sin(x)$

- 求sin(x)的最大值与最小值，令原函数的一阶导数$y=cos(x)=0$,这样可以得到sin(x)的极值，此时$x=\frac{\pi}{2}$
- 在区间$[0,\pi]$,二阶导数小于0，表示在这个区间上，表示斜率持续降低，同时由于二阶导数小于0表示函数为凹函数，在$x=\frac{\pi}{2}$点取得的是函数的最大值。
- 在$x=\pi$时，二阶导数=0，表示这是函数的拐点，经过拐点函数有凹函数变为了凸函数

### 例子：从家到MIT的最短时间

```python
y = np.arange(0,100,1)
x = np.zeros_like(y)+50
plt.plot(x,y)
plt.text(50,80,"b")

x_1 = np.arange(40,51,1)
y_1 = np.zeros_like(x_1)+40
plt.plot(x_1, y_1)

y_2 = 2*x_1-40
plt.plot(x_1, y_2)
plt.text(44,38,"a")
plt.text(50,50,"x")
plt.text(46,50,"sqrt(x²+a²)")
```

- 假设汽车在b行驶速度为60，在a(或$\sqrt{(x^2+a^2)}$))行驶速度为30，那么最短时间：$TIME = \frac{b-x}{60} + \frac{\sqrt{(x^2+a^2)}}{30}$，求x等于多少时行驶时间最短？
- 解：
>- 求TIME的一阶导数$TIME' = -\frac{1}{60}+\frac{1}{30}\frac{1}{2}\frac{2x}{\sqrt{(a^2+x^2)}}$
>- 令$TIME'=0$可以得到$x=\frac{a}{\sqrt{3}}$时取的极值
