# 微积分简介
- 微积分是研究两个函数之间关系的学科(函数都是成对的)
- 微积分的作用是通过函数一（微风函数）求解另函数二（积分函数），这两个函数包含相同的信息，因此微积分分为微分学和积分学

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 例子
- 以速度、加速度、时间、距离举例

## 案例一
> - 距离`f(x)=v * t`; `v=40`，两个函数如下：

```python
# 速度
v = 40
# 时间
t = np.arange(1,100,1)
# 距离
fx = v*t
```

```python
# 用于画图
plot_v = np.ones_like(t)*40
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2, nrows=1,figsize=(20, 10))
ax1.plot(t, fx, linewidth=4)
ax1.set_title("Function f = v*t", fontsize=40)
ax1.set_xlabel("Time", fontsize=20)
ax1.set_ylabel("Distance", fontsize=20)
ax1.set_xlim(1,100)
ax1.set_ylim(1,4000)

ax2.plot(t, plot_v, linewidth=4)
ax2.set_title("Function V = df/dt", fontsize=40)
ax2.set_xlabel("Time", fontsize=20)
ax2.set_ylabel("V", fontsize=20)
ax2.set_xlim(1,100)
ax2.set_ylim(1,100)
plt.show()
```

## 案例二
- 假设加速度恒定为a，初始速度为0， 距离为f，求在时间t时距离f的值
> - 加速度与速度的计算公式如下：$v = a * t$
> - 通过函数v可以求出函数f：

 - $f = v*t$

 - $v_0 = 0 + a*0$

 - $v_1 = 0 + a*t$

 - $f = \frac{v_0+v_1}{2}*t = \frac{1}{2} at^2$

```python
# 速度与时间之间的关系
a = 10
t = np.arange(0,100,1)
v = a*t
f = 0.5*a*np.square(t)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax1.plot(t, v, linewidth=4)
ax1.set_title("Function v = df/dt", fontsize=40)
ax1.set_xlim(1,100)
ax1.set_ylim(1,1100)
ax1.set_xlabel("Time", fontsize=20)
ax1.set_ylabel("V", fontsize=20)

ax2.plot(t, f, linewidth=4)
ax2.set_title("Function f = 1/2*a*t^2", fontsize=40)
ax2.set_xlim(1,20)
ax2.set_ylim(1,1100)
ax2.set_xlabel("Time", fontsize=20)
ax2.set_ylabel("f", fontsize=20)
```
